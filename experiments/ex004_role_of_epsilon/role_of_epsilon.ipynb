```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams.update({'font.size': 14})

import numpy as np
import numpy.linalg as la
MEPS = np.finfo(float).eps

from scipy.optimize import minimize_scalar

import pandas as pd
```

```python
def functional(eps, dist_mat, rbf, P=None, target_cond=10**12):
    A = rbf(dist_mat, eps)
    if not P is None:
        return np.log(la.cond(A)/target_cond)**2
        k = len(P[0])
        A = np.block([[A, P],[P.T, np.zeros((k,k))]])
    return np.log(la.cond(A)/target_cond)**2 

def optimize_eps(rbf, dist_mat, P=None, target_cond=10**12):
    n = dist_mat.shape[0]
    eps_guess = 1/np.min(dist_mat+np.diag([1]*n))
    optimization_result = minimize_scalar(functional, 
            bracket=[eps_guess/2, eps_guess*2],
            args=(dist_mat, rbf, P, target_cond))
    return optimization_result['x']

def rbf_interp(xs, fs, zs, rbf, deg=-1, shape=True, print_cond=False, ret_eps=False, eps=None):
    dist_mat = np.abs(np.subtract.outer(xs,xs))
    n = len(xs)
    if deg == -1:
        P=None
    else:
        P = xs.reshape((n, 1)) ** np.arange(deg+1)
    if eps is None and shape==True:
        eps = optimize_eps(rbf, dist_mat, P=P)
    A = rbf(dist_mat, eps)
    if deg != -1:
        A = np.block([[A, P],[P.T, np.zeros((deg+1,deg+1))]])
    if print_cond:
        print('log Cond(A)= %g' % ( np.log(la.cond(A))/np.log(10) ) )
    cs = la.solve(A, np.block([fs, np.zeros(deg+1)]))
    A = rbf(np.abs(np.subtract.outer(zs,xs)), eps)
    if deg != -1:
        P = zs.reshape((len(zs), 1)) ** np.arange(deg+1)
        A = np.block([A, P])
    ret = A@cs
    if ret_eps:
        return ret, eps
    return ret
```

```python
def rbf(r, eps):
    return np.exp(-(eps*r)**2)

a, b = -1, 1
def foo(x):
    #return (x-0)**2 * x*(x-1)
    #return np.sin(2*np.pi*x)+1
    return np.sin(np.pi* 4*x**2)

xs = np.linspace(a, b, 1000)
plt.figure(figsize=(12,8))
plt.plot(xs, foo(xs),'b-')
plt.show()
```

```python
n, m = 20, 1000

xs = np.linspace(a, b, n)
zs = np.linspace(a, b, m)
us = rbf_interp(xs, foo(xs), zs, rbf, deg=4)

err = la.norm(us - foo(zs))/la.norm(foo(zs))

print('error: %g' % err)

plt.figure(figsize=(12,8))
plt.plot(xs, foo(xs),'bo')
plt.plot(zs, foo(zs), 'b-')
plt.plot(zs, us, 'r--')
plt.show()
```

```python
degs = range(-1,3)
ns = [256*2**i for i in range(5)]
print(ns)

errors = []
epsilons = []
m = 4000
zs = np.linspace(a, b, m)

for d in degs:
    errors += [[]]
    for n in ns[::-1]:
        print('calculating deg=%d \tn=%d' % (d,n))
        xs = np.linspace(a, b, n)
        if n == ns[-1]:
            us, eps = rbf_interp(xs, foo(xs), zs, rbf, deg=d, ret_eps=True)
            epsilons += [eps]
        else:
            us = rbf_interp(xs, foo(xs), zs, rbf, deg=4, eps=eps)

        err = la.norm(us - foo(zs))/la.norm(foo(zs))
        errors[-1] += [err]
```

```python
plt.figure(figsize=(12,8))
for errs in errors:
    plt.loglog(ns, errs[::-1],'o--')
plt.xticks(ns, ['%d'%n for n in ns])
plt.legend(['deg=%d'%d for d in degs])
plt.show()
```

```python
epsilons = [.1, .2, .5, 1, 2, 4, 10]
ns = [8*2**i for i in range(6)]
print(ns)
#d = 0
degs = range(-1, 4)

deg_errors = []
m = 4000
zs = np.linspace(a, b, m)

for d in degs:
    deg_errors += []
    errors = []
    for eps in epsilons:
        errors += [[]]
        for n in ns[::-1]:
            xs = np.linspace(a, b, n)
            us = rbf_interp(xs, foo(xs), zs, rbf, deg=d, eps=eps)

            err = la.norm(us - foo(zs))/la.norm(foo(zs))
            errors[-1] += [err]
    deg_errors += [errors]
```

```python
for d, errors in zip(degs, deg_errors):
    plt.figure(figsize=(12,8))
    for errs in errors:
        plt.loglog(ns, errs[::-1],'--')
    plt.legend(['$\\varepsilon = %.2f$' % eps for eps in epsilons])
    plt.title('deg = %d' % d)
    plt.xticks(ns, ns)
    plt.show()
```

```python
a, b = -1, 1
def foo(x):
    return np.sin(np.pi* 2*x) + (x-.1)*(x+.2)*(x-.5)*(x+.8)

xs = np.linspace(a, b, 1000)
plt.figure(figsize=(12,8))
plt.plot(xs, foo(xs),'b-')
plt.show()
```

```python
n, m = 30, 1000
deg = 3

xs = np.linspace(a, b, n)
zs = np.linspace(a, b, m)

us = rbf_interp(xs, foo(xs), zs, rbf, deg=deg)
err = la.norm(us - foo(zs))/la.norm(foo(zs))
print('error: %g' % err)

plt.figure(figsize=(12,8))
plt.plot(xs, foo(xs),'bo')
plt.plot(zs, foo(zs), 'b-')
plt.plot(zs, us, 'r--')
plt.show()
```

```python
degs = range(-1,20)
N = 200
m = 1000

zs = np.linspace(a, b, m)
for d in degs:
    k = d+1
    xs = np.linspace(a, b, N+k)
    len1 = len(xs)
    us = rbf_interp(xs, foo(xs), zs, rbf, deg=-1)
    err = la.norm(us - foo(zs))/la.norm(foo(zs))
    #print('unaugmented error: \t%g' % err )
    err_standard = err
    
    xs = np.linspace(a, b, N)
    len2 = len(xs)
    us = rbf_interp(xs, foo(xs), zs, rbf, deg=d)
    err = la.norm(us - foo(zs))/la.norm(foo(zs))
    #print('augmented error: \t%g' % err )
    err_aug = err
    
    print('\ndeg = %d' % d)
    print('ratio: %g' % (err_aug/err_standard))
    print('error augmented: %g' % err_aug)
    
        
```
