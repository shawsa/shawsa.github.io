```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(12, 5))
def foo(x):
    return np.floor(x).astype(np.int)%2

def goo(x):
    return np.floor(np.sqrt(2)*x).astype(np.int)%2
    
plt.plot(x, foo(x)+goo(x))
```

```python
fig = plt.figure(figsize=(12, 5))
def foo(x):
    return x - np.floor(x)

def goo(x):
    return np.sqrt(2)*x - np.floor(np.sqrt(2)*x)
    
plt.plot(x, foo(x)+goo(x))
```
