```python
n = 10
R, r = 5, 1
nodes = np.meshgrid(np.linspace(0,1,n*int(R/r)), np.linspace(0,1,n*int(r/R+1)))
nodes = np.block([[nodes[0].flatten()],[ nodes[1].flatten()]]).T
```

```python
fig = plt.figure(figsize=(4,4))
plt.plot(nodes[:,0], nodes[:,1], 'b.')
```

```python
def square_to_torus(nodes, R=1, r=.5):
    theta = (nodes[:,0]-.5 )*2*np.pi
    phi = (nodes[:,1]-.5 )*2*np.pi
    z = np.sin(phi)*r
    x = np.cos(theta)*(R + np.cos(phi)*r)
    y = np.sin(theta)*(R + np.cos(phi)*r)
    return x, y, z
```

```python
x, y, z = square_to_torus(nodes, R=5, r=1)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x, y, z, '.', color='b')

ax.set_xlim(-6,6)
ax.set_ylim(-6,6)
ax.set_zlim(-6,6)

plt.show()
```
