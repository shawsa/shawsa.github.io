# Iterated Differentiation
Some basic testing of the technique of iterated differentiation in one dimension.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

import scipy.linalg as la
```

```python
# Gaussian
def rbf1(r):
    return np.exp(-(epsilon*r)**2)
def d_rbf1(r):
    return -2 * epsilon**2 * r * np.exp(-(epsilon*r)**2)
    
# PHS
def rbf2(r):
    return r**7
def d_rbf2(r):
    return 7*r**6
```

```python
n = 100
stencil_size = 15
epsilon = 1

rbf, d_rbf = rbf1, d_rbf1


def foo(x):
    return np.sin(3*np.pi*x)

def d_foo(x):
    return 3*np.pi * np.cos(3*np.pi * x)

def d2_foo(x):
    return -(3*np.pi)**2 * np.sin(3*np.pi * x)
```

```python
xs = np.linspace(0,1,n)
ys = foo(xs)

xs_plot = np.linspace(0,1, 1000)
ys_plot = foo(xs_plot)
plt.figure(figsize=(8,8))
plt.plot(xs_plot, ys_plot, 'b-')
plt.plot(xs, ys, 'ro')
```

# Approximate First Derivative at a Point

```python
# Stencil
index = int(2*n/5)
lower = index - (stencil_size-1)//2
upper = index + (stencil_size-1)//2

nodes = xs[lower:upper+1]

plt.figure(figsize=(8,8))
plt.plot(xs_plot, ys_plot, 'b-')
plt.plot(xs, ys, 'ro')
plt.plot(nodes, foo(nodes), 'go')
plt.plot(xs[index], foo(xs[index]), 'yo')
```

```python
A = np.array([[rbf(np.abs(x-x_0)) for x_0 in nodes] for x in nodes])
rhs = np.array([d_rbf(np.abs(xs[index] - x)) * np.sign(xs[index] - x) for x in nodes])
print(np.linalg.cond(A))
```

```python
W = la.solve(A, rhs)
plt.figure(figsize=(8,8))
plt.plot(xs_plot, d_foo(xs_plot), 'b-')
plt.plot(xs[index], W@foo(nodes), 'ro')
```

# Approximate derivative at all points in stencil

```python
rhs = np.array([[d_rbf(np.abs(x_0 - x)) * np.sign(x_0 - x) for x_0 in nodes] for x in nodes])
W = la.solve(A, rhs).T
plt.figure(figsize=(8,8))
plt.plot(xs_plot, d_foo(xs_plot), 'b-')
plt.plot(nodes, W@foo(nodes), 'ro')
```

# Approximate at all points in stencil.

```python
plt.figure(figsize=(8,8))
plt.plot(xs_plot, d2_foo(xs_plot), 'b-')
plt.plot(nodes, W@W@foo(nodes), 'ro')
```
