```python
foo = lambda x: np.sqrt(1-x**2)
dfoo = lambda x: -x/foo(x)

x0 = .5
win_width = .5

xsc = np.linspace(x0-.4, x0+.4, 7)
ysc = foo(xsc)

y0 = foo(x0)
m = dfoo(x0)

xsp = (ysc - y0 + m*x0 + 1/m*xsc)/(m+1/m)
ysp = dfoo(x0)*(xsp-x0)+y0

xs = np.linspace(0, 1)

fig = plt.figure(figsize=(6,6))
plt.plot(xs, foo(xs), 'k-')
plt.plot(xs, dfoo(x0)*(xs-x0)+y0, 'k-')
for (x, y) in zip(zip(xsp, xsc), zip(ysp, ysc)):
    plt.plot(x, y, 'ko--')
# plt.plot(x_ns, foo(x_ns), 'ko')
# plt.plot(xsp, ysp, 'ko')

plt.xlim(x0-win_width, x0+win_width)
plt.ylim(y0-win_width, y0+win_width-.2)
plt.axis('off')
# plt.title('Figure 1', y=0)
plt.show()
```
