```python
import matplotlib.pyplot as plt
import sympy as sym
sym.SYMPY_DEBUG = True
my_modules = [{'Heaviside': lambda x: np.heaviside(x,0.5)}, 'numpy']
import numpy as np

x, y = sym.symbols('x y', real=True)
x0 = sym.symbols('x0', real=True)
x1 = sym.symbols('x1', real=True, positive=True)
expr = sym.Heaviside(x-y)*sym.exp(y-x)
F = sym.integrate(expr, (y,-x0, 0))
G = sym.integrate(expr, (y,-x1, 0))

display(F)
display(G)

my_modules = [{'Heaviside': lambda x: np.heaviside(x,0.5)}, 'numpy']
Fnum = sym.lambdify(x, F.subs(x0,1), modules=my_modules)
Gnum = sym.lambdify(x, G.subs(x1,1), modules=my_modules)
plt.figure(figsize=(10,8))
xs = np.linspace(-2,2,201)
plt.plot(xs, Fnum(xs), 'b-')
plt.plot(xs, Gnum(xs), 'm-')
plt.ylim(-.1, 1.1)
plt.show()
```
